```python
%matplotlib inline
import pandas as pd
import seaborn as sns; sns.set(style="ticks", color_codes=True)
import matplotlib.pyplot as plt
sns.set_style("white")
```

```python
df = pd.read_csv('aopc3.csv')
df
```

```python
df['architecture_seq'] = df.apply(lambda row: '%s_seq-%d' % (row['architecture'], row['seq']), axis=1)
```

```python
def plot_dataset(name):
    plt.suptitle('Dataset %s' % name)
    sns.factorplot(x="method", y="aopc_area", col="architecture_seq", data=df[df.dataset==name],
               col_wrap=4, markers=['o',  '^', '*', 'x'], ci=None, palette="muted", size=5,
               scatter_kws={"s": 50, "alpha": 1}, fit_reg=False)
   
plot_dataset('mnist')
```

```python
plot_dataset('fashion-mnist')
```
